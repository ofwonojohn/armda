```python
# Make a copy of the original dataset for imputation
df_imputed = df.copy()

# Handle missing values
df_imputed['Age'].fillna(df_imputed['Age'].median(), inplace=True)
df_imputed['Income'].fillna(df_imputed['Income'].median(), inplace=True)
df_imputed['CreditScore'].fillna(df_imputed['CreditScore'].mean(), inplace=True)
df_imputed['LoanAmount'].fillna(df_imputed['LoanAmount'].median(), inplace=True)

# Check if there are any remaining missing values
print("Missing values after imputation:")
df_imputed.isnull().sum()
```
